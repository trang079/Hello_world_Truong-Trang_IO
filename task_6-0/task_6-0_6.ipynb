```python
import pandas as pd

# --- РАСЧЕТ КОЭФФИЦИЕНТА ВАРИАЦИИ ---

# 1. Чтение файла с данными
df = pd.read_csv('wild_boars.csv')

# 2. Расчет коэффициента вариации для веса (weight_kg)
# Формула: (Стандартное отклонение / Среднее арифметическое) * 100
cv_weight = (df['weight_kg'].std() / df['weight_kg'].mean()) * 100

# 3. Вывод результата на экран
print(f"Coefficient of variation is {cv_weight:.2f} %")
```
