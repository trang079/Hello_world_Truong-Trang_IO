```python
import pandas as pd

# 1. Чтение файла с данными
df = pd.read_csv('wild_boars.csv')

# 2. Расчет средних значений по всем числовым столбцам
mean_values = df.mean(numeric_only=True)

# 3. Сохранение результатов в отдельный CSV файл
mean_values.to_csv('mean_results.csv', header=['mean_value'])

# 4. Вывод результатов на экран для проверки
print("--- СРЕДНИЕ ЗНАЧЕНИЯ ПО ВСЕМ СТОЛБЦАМ ---")
print(mean_values)
print("\nРезультаты успешно сохранены в файл 'mean_results.csv'")
```
