```python
import pandas as pd

# --- ЗАДАНИЕ 7 ---

# 1. Чтение файла с данными
df = pd.read_csv('wild_boars.csv')

# Выбираем только числовые столбцы
numeric_df = df.select_dtypes(include=['number'])

# 2. Расчет статистических показателей
variance = numeric_df.var()
std_deviation = numeric_df.std()
coef_variation = (std_deviation / numeric_df.mean()) * 100

# Создаем итоговую таблицу для сохранения
results = pd.DataFrame({
    'variance': variance,
    'std_deviation': std_deviation,
    'coef_variation_percent': coef_variation
})

# 3. Сохранение результатов в отдельный CSV файл
results.to_csv('variance_std_cv_results.csv')

# 4. Вывод результатов на экран для проверки
print("--- РЕЗУЛЬТАТЫ РАСЧЕТА ДЛЯ ВСЕХ ЧИСЛОВЫХ СТОЛБЦАМ ---")
print(results)
print("\nРезультаты успешно сохранены в файл 'variance_std_cv_results.csv'")
```
