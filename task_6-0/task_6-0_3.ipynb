```python
import pandas as pd

# --- ЗАДАНИЕ 3 ---

# 1. Чтение файла с данными
df = pd.read_csv('wild_boars.csv')

# 2. Расчет медианных значений по всем числовым столбцам
median_values = df.median(numeric_only=True)

# 3. Сохранение результатов в отдельный CSV файл
# Каждая строка автоматически начнется с названия столбца (параметра)
median_values.to_csv('median_results.csv', header=['median_value'])

# 4. Вывод результатов на экран для проверки
print("--- МЕДИАННЫЕ ЗНАЧЕНИЯ ПО ВСЕМ СТОЛБЦАМ ---")
print(median_values)
print("\nРезультаты успешно сохранены в файл 'median_results.csv'")
```
