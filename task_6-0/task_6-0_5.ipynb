```python
import pandas as pd

# --- ЗАДАНИЕ 5 ---

# 1. Чтение файла с данными
df = pd.read_csv('wild_boars.csv')

# Выбираем только числовые столбцы для расчета процентилей
numeric_df = df.select_dtypes(include=['number'])

# 2. Определение списка процентилей (25%, 50%, 75%, 90%, 95%, 100%)
quantiles_list = [0.25, 0.50, 0.75, 0.90, 0.95, 1.00]

# Расчет процентилей по всем числовым колонкам
percentiles_df = numeric_df.quantile(quantiles_list)

# Переименовываем индексы для красивого вывода, как в примере
percentiles_df.index = [
    'Percentile 25 (Q1)', 
    'Median 50 (Q2)', 
    'Percentile 75 (Q3)', 
    'Percentile 90', 
    'Percentile 95', 
    'Max'
]

# 3. Сохранение результатов в отдельный CSV файл
percentiles_df.to_csv('percentile_results.csv')

# 4. Вывод результатов на экран в текстовом формате для проверки
print("--- РЕЗУЛЬТАТЫ РАЗБИВКИ ПО ПРОЦЕНТИЛЯМ ---")
for col in percentiles_df.columns:
    print(f"\n{col.capitalize()}:")
    for idx, val in percentiles_df[col].items():
        print(f"  {idx}: {val:.1f}")

print("\nРезультаты успешно сохранены в файл 'percentile_results.csv'")
```
