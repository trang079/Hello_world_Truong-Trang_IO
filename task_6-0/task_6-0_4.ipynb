```python
import pandas as pd

# --- ЗАДАНИЕ 4 ---

# 1. Чтение файла с данными
df = pd.read_csv('wild_boars.csv')

# 2. Расчет моды для каждого столбца и выбор первого значения, если мод несколько
# Метод .mode() работает как с числами, так и с текстом (категориальными данными)
mode_values = df.mode().iloc[0]

# 3. Сохранение результатов в отдельный CSV файл
# Каждая строка автоматически начнется с названия столбца (параметра)
mode_values.to_csv('mode_results.csv', header=['mode_value'])

# 4. Вывод результатов на экран для проверки
print("--- МОДАЛЬНЫЕ ЗНАЧЕНИЯ ПО ВСЕМ СТОЛБЦАМ ---")
print(mode_values)
print("\nРезультаты успешно сохранены в файл 'mode_results.csv'")
```
