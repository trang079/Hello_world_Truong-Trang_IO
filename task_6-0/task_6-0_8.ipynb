```python
import pandas as pd

# --- ЗАДАНИЕ 8 ---

# 1. Чтение файла с данными
df = pd.read_csv('wild_boars.csv')

# 2. Функция для расчета коэффициента вариации (CV) для определенного пола
def calculate_cv(gender_type):
    gender_df = df[df['gender'] == gender_type]
    # Формула: (Стандартное отклонение / Среднее арифметическое) * 100
    cv = (gender_df['tusk_length'].std() / gender_df['tusk_length'].mean()) * 100
    return cv

# Расчет CV отдельно для самцов (male) и самок (female)
cv_male = calculate_cv('male')
cv_female = calculate_cv('female')

# Создание DataFrame для хранения результатов
results = pd.DataFrame({
    'gender': ['male', 'female'],
    'tusk_length_cv_percent': [cv_male, cv_female]
})
results.set_index('gender', inplace=True)

# 3. Сохранение результатов в отдельный CSV файл
results.to_csv('tusk_cv_results.csv')

# 4. Вывод результатов на экран для проверки
print("--- КОЭФФИЦИЕНТ ВАРИАЦИИ ДЛЯ ДЛИНЫ КЛЫКОВ ПО ПОЛУ ---")
print(results)
print("\nРезультаты успешно сохранены в файл 'tusk_cv_results.csv'")
```
